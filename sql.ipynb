```python
db = sqlite3.connect("data/pokemon.db")
```

```python
import json
import os
with open(os.path.join("./data/", "pokemon.json"), newline='', encoding='UTF-8') as jsonfile:
    pokemon = json.load(jsonfile)
    jsonfile.close()
status = { }
for user in pokemon.keys():
    id=user
    name=pokemon[user][0]
    hisstatus=pokemon[user][1]
    atype=hisstatus[7]
    btype=""
    level=hisstatus[6]
    hp=hisstatus[0]
    speed=hisstatus[5]
    atk=hisstatus[1]
    defv=hisstatus[2]
    spatk=hisstatus[3]
    spdef=hisstatus[4]
    money=0
    exp=0
    
    cmd = "insert into pokemon(id	,name,typea,typeb,level,hp,	speed,	atk,	def,	spatk,	spdef,	money,	exp) values('{}','{}','{}','{}','{}','{}','{}','{}','{}','{}','{}','{}','{}')".format(id,name,atype,btype,level,hp,speed,atk,defv,spatk,spdef,money,exp)
    db.execute(cmd)
    db.commit()
```
